```python
import os
import pandas as pd
from funcs import extract_run_info, get_n_jobs
```

```python
dr = '../Exp/Job_Logs'
files = os.listdir(dr)
files = [file for file in files if '.err' in file]
print(len(files), 'Error Files found!')

errors = pd.DataFrame(columns=['model', 'parcel', 'n_jobs', 'cnt'])

for file in files: 
    full_path = os.path.join(dr, file)

    # Open file
    with open(full_path, 'r') as f:
        txt = f.readlines()
    
    error_str = ' DUE TO TIME LIMIT '
    is_error = False
    for line in txt:
        if error_str in line:
            is_error = True
            continue
    
    out_file = full_path.replace('.err', '.out')
    if is_error and os.path.exists(out_file):
        
        with open(out_file, 'r') as f:
            txt = f.readlines()
            
        run_info = extract_run_info(txt)
        
        n_strip = run_info.split('---')
        parcel = n_strip[0]
        model = n_strip[1]
        
        n_jobs = get_n_jobs(file)
        errors.loc[len(errors)] = [model, parcel, n_jobs, 1]
         
counts_df = errors.groupby(['model', 'parcel', 'n_jobs']).agg('count')
```
