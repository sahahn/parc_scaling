```python
from nilearn.plotting import plot_surf_stat_map, plot_surf_roi
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rh(mode='curv'):

    data_loc = '../data/abcd_structural/' + mode + '/INV0B0HVEYX.npy'
    data = np.load(data_loc)
    rh_data = data[len(data) // 2:]
    
    return rh_data

def rh_rois(name='schaefer_100'):
            
    data_loc = '../parcels/' + name + '.npy'
    data = np.load(data_loc)
    rh_data = data[len(data) // 2:]
    
    return rh_data

rh_mesh = '/home/sage/Neuro_Plotting/data/32k_fs_LR/surf/rh.inflated'
```

```python
plot_surf_stat_map(rh_mesh, rh('curv'),
                   hemi='right', colorbar=False,
                   cmap='blue_orange')

plt.title('Curvature', fontsize=28)
plt.savefig('Figures/curv.png', dpi=300, transparent=True, bbox_inches='tight')
```

```python
plot_surf_stat_map(rh_mesh, rh('thick'),
                   hemi='right', colorbar=False,
                   cmap='ocean_hot')

plt.title('Thickness', fontsize=28)
plt.savefig('Figures/thick.png', dpi=300, transparent=True, bbox_inches='tight')
```

```python
plot_surf_stat_map(rh_mesh, rh('sulc'),
                   hemi='right', colorbar=False,
                   cmap='cold_hot')

plt.title('Sulcal Depth', fontsize=28)
plt.savefig('Figures/sulc.png', dpi=300, transparent=True, bbox_inches='tight')
```

```python
plot_surf_stat_map(rh_mesh, rh('myelin'),
                   hemi='right', colorbar=False, symmetric_cbar=False,
                   cmap='hot_black_bone_r', vmax=6)

plt.title('Myelination', fontsize=28)
plt.savefig('Figures/myelin.png', dpi=300, transparent=True, bbox_inches='tight')
```
